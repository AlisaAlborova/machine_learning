```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
df1=gpd.read_file("altaiskii-krai.geojson")
```

```python
df1.head(3)
```

```python
zzz=pd.json_normalize(df1['point'])
```

```python
df1['lat']=zzz['lat']
df1['long']=zzz['long']
```

```python
df1.drop('point',axis=1,inplace=True)
df1.head(3)
```

```python
df2=gpd.read_file("amurskaia-oblast.geojson")
```

```python
zzz1=pd.json_normalize(df2['point'])
```

```python
df2['lat']=zzz1['lat']
df2['long']=zzz1['long']
```

```python
df2.drop('point',axis=1,inplace=True)
df2.head(3)
```

```python
frames = [df1, df2]
```

```python
result = pd.concat(frames)
```
